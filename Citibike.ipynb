Downloading May 2014 citibike data. This data set contains the start and stop time and location for ever cibike user trip (it also includes the user age and gender, but I didn't use those). In order to look at how many bikes are added to or taken away from a station, the data is rearranged so that each row has the station, time, and a +1 if the bike is being returned to the station and a -1 is the bike is being checked out.

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

df = pd.read_csv('citibike_data/2014-05-CitiBikeTripData.csv') #reading in file to dataframe
startdf=df[['starttime','start station id','start station name']] # contains starting stations and times
startdf.columns = ['time','station id','station name']
startdf['bike']=-1 # start of trip removes a bike from station
enddf=df[['stoptime','end station id','end station name']] # containts ending stations and times
enddf.columns = ['time','station id','station name']
enddf['bike']=1 #end of trip adds a bike to the station
stationdf = pd.concat([startdf,enddf],ignore_index = True) #contains starting and ending stations
stationdf = stationdf.sort('time').reset_index(drop=True)
stationdf['datetime']=pd.to_datetime(stationdf.time) # converting time as string into datetime
stationdf.drop('time',axis=1,inplace=True) # removing time as string column
stationdf.set_index('datetime',inplace=True) # datetime is now the index
stationdf['dayofweek']=stationdf.index.dayofweek #adding column with day-of-week 0=Monday,...
stationdf['date']=stationdf.index.date
```

Here is the new dataframe created. It is indexed by the data/time the bike was checked out/in. It contains the station name and id. The column 'bike' tells us if a bike was checked out or in (each row represents one bike). There is also a column for the date and the 'day of week' where 0 is Monday and 6 is Sunday.

Using 'groupby' we can add up all the bikes checked into and out of the entire system for everyday of May. It would be nice if the plot could also show the day of week [M,T,W,...] or at least indicate where the weekends are.

```python
dailyBike=stationdf.groupby('date').sum().bike
dailyBike.plot(kind='bar')
plt.grid(True,which='major')
#plt.xticks(i for i in pd.date_range('5-1-2014',periods=31,freq='D'))
plt.xlabel('date')
plt.ylabel('bikes in minus bikes out')
plt.title('Citibike - May 2014: Daily number of bikes checked in')
#plt.minorticks_on()
plt.show()

```
