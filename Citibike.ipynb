## NYC's Bike Share: Following the Bikes

### Reading in the trip histories data, downloaded from the Citibike Website.

```python
%matplotlib inline
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import citibikefncs as cb
```

```python
#reading in data
#filenumbers refer to month
df4 = pd.read_csv('data/2014-04 - Citi Bike trip data.csv') 
df5 = pd.read_csv('data/2014-05 - Citi Bike trip data.csv') 
df6 = pd.read_csv('data/2014-06 - Citi Bike trip data.csv') 
df7 = pd.read_csv('data/2014-07 - Citi Bike trip data.csv') 
df8 = pd.read_csv('data/2014-08 - Citi Bike trip data.csv') 

dataframes=[df4,df5,df6,df7,df8]

for df in dataframes:
    df['starttime']=pd.to_datetime(df['starttime']) #converting time strings into datetime format var
    df['stoptime']=pd.to_datetime(df['stoptime'])   #converting time strings into datetime format var
    
dfall=pd.concat(dataframes)
```

```python
df4.tail()
```

```python
#tapping into the citibike json feed
#also gives the current number of available docks/bikes per station (not using that info)
import urllib, json
from pandas.io.json import json_normalize
url = "http://citibikenyc.com/stations/json"

response = urllib.urlopen(url)
data = json.loads(response.read())
stationdf=json_normalize(data['stationBeanList'])
updateTime = data['executionTime']
stationdf['time']=updateTime
stationdf.set_index(['id'],drop=False,inplace=True)

#printing out station list with attributes to csv file
stationinfo=stationdf[['stationName','totalDocks','latitude','longitude']]
#stationinfo.to_csv('stationinfo.csv')

#including new station ids so that they are labeled in order starting from 0
stationinfo['new id']=stationinfo.reset_index().index
```

### Bike share stats

```python
print 'Bike share stats:'
print '\t total number of stations: ',stationinfo.index.size
print '\t number of docks per station: max: ', stationinfo.totalDocks.max(), \
'\t average: ', int(round(stationinfo.totalDocks.mean()))
print '\t system capacity - total docks: ',stationinfo.totalDocks.sum()
print '\t total bikes:', df8.groupby('bikeid').count().count()[0]
```

```python
#calculates whereisbike for each month by using where the bike was last dropped off
#by a rider. The bike stays at that station until it is picked up by another rider.
#whereisbike columns give bike id, there is one row for each hour of the month.

#this cell takes a couple minutes to run, time it:
then=dt.datetime.now()

bikedf4=cb.indexByBikeidTime(df4)
nullstartrow=pd.Series(index=bikedf4.index.levels[0])
whereisbike4=cb.tripToStation(bikedf4,dt.datetime(2014,4,1,0),1,nullstartrow)

startrow=whereisbike4.loc[dt.datetime(2014,4,30,0)]
bikedf5=cb.indexByBikeidTime(df5)
whereisbike5=cb.tripToStation(bikedf5,dt.datetime(2014,5,1,0),1,startrow)

startrow=whereisbike5.loc[dt.datetime(2014,5,31,0)]
bikedf6=cb.indexByBikeidTime(df6)
whereisbike6=cb.tripToStation(bikedf6,dt.datetime(2014,6,1,0),1,startrow)

startrow=whereisbike6.loc[dt.datetime(2014,6,30,0)]
bikedf7=cb.indexByBikeidTime(df7)
whereisbike7=cb.tripToStation(bikedf7,dt.datetime(2014,7,1,0),1,startrow)

startrow=whereisbike7.loc[dt.datetime(2014,7,31,0)]
bikedf8=cb.indexByBikeidTime(df8)
whereisbike8=cb.tripToStation(bikedf8,dt.datetime(2014,8,1,0),1,startrow)

now=dt.datetime.now()
print now-then
```

### How long is a bike at a station?

Because the trip histories data gives bike id, we can track the bikes and figure out their downtime.

```python
stationstays, bikestays = cb.bikeStationStays(whereisbike8,stationinfo)
```

```python
#single bike - distribution of hours idle
i=0
colname=bikestays.columns[i]
bh=bikestays[[colname]]
bh.rename(columns={colname:'freq'},inplace=True)
bh['hours']=bh.index
nm=float(bh.freq.sum())
bh['norm']=bh.freq.apply(lambda x: x/nm)
bh.plot(x='hours',y='norm',kind='scatter',alpha=.50)
plt.title('single bike - distribution of hours idle',fontsize=14)
plt.ylabel('freq')
```

```python
#looking at the distriubtion of stays from all bikes at all stations
#remove stations id -10 (these are bikes that are out) and do a value counts on the remaining
vc=stationstays[stationstays['station new id']!=-10]['hours'].value_counts()
norm=float(vc.sum())
normed=pd.DataFrame(index=vc.index)
normed['counts']=vc.apply(lambda x: x/norm)
normed['hours']=normed.index
normed.sort('hours',inplace=True)
normed.plot(x='hours', y='counts',kind='line')
#plt.plot(decay(t,N,beta))
plt.xlim(0,48)
#plt.ylim(.0001,.8)
plt.yscale('log')
plt.title('all bikes - distribution of time spent idle at stations')
plt.xlabel('hours')
plt.ylabel('normalized frequency')
```

### Do any stations stand out as having more bikes idle?

```python
#looking at stations, plot gives a visualization of number and length of bike stays
#each dot is one stay, the x axis is all the stations in the system

stationstays.plot(x='station new id',y='hours',kind='scatter',alpha=.01)
plt.xlabel('stations')
plt.ylabel('hours at station')
plt.title('Visualization of number and length of bike stays.\n Each dot is one stay.')
plt.ylim(0,100)
#plt.xlim(275,300)
```

### Rebalancing

```python
#when are the bikes being rebalanced?
#determined by when a bike is picked up at a different station
#from where it was last returned to
#addbike gives the addition of bikes (ie to the station it was 
#picked up from, using the time it was picked up as a proxy)
#removebike gives the number of bikes removed from station, using
#the time it was dropped off as a proxy.

then=dt.datetime.now()
addbike, removebike = cb.rebalanced(bikedf5)
now=dt.datetime.now()
print now-then
```

```python
stationadds=cb.stationFill(addbike,stationinfo)
stationrm=cb.stationFill(removebike,stationinfo)
stationbal=stationadds-stationrm

#weekday averages
addsavg=cb.weekDayAvg(stationadds)
rmavg=cb.weekDayAvg(stationrm)
```

```python
stno=521
stname=stationinfo.loc[stno]['stationName']
plt.subplot(221)
sfavg[stno].plot(legend=False)
plt.ylabel('total bikes',fontsize=14)
plt.xlabel('',fontsize=14)
plt.title(stname,fontsize=14)

plt.subplot(223)
addsavg[stno].plot()
rmavg[stno].apply(lambda x: -x).plot(color='r')
plt.ylabel('bikes added/removed',fontsize=14)
plt.xlabel('hour of the day',fontsize=14)
#plt.title('Rebalanced: bikes added and removed\n'+stname,fontsize=14)
#fig=plt.gcf()
#fig.set_size_inches(5,8)

stno=465
stname=stationinfo.loc[stno]['stationName']
plt.subplot(222)
sfavg[stno].plot(legend=False)
plt.ylabel('total bikes',fontsize=14)
plt.xlabel('',fontsize=14)
plt.title(stname,fontsize=14)

plt.subplot(224)
addsavg[stno].plot()
rmavg[stno].apply(lambda x: -x).plot(color='r')
plt.ylabel('bikes added/removed',fontsize=14)
plt.xlabel('hour of the day',fontsize=14)
#plt.title('Rebalanced: bikes added and removed\n'+stname,fontsize=14)
fig=plt.gcf()
fig.set_size_inches(12,8)
```
