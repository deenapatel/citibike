Read in Citibike May 2014 trip histories data (downloaded from http://www.citibikenyc.com/system-data). This data contains all trips, including their origin/destination information. It does not include any rebalancing (redistribution) of bikes that Citibike may have done, so we don't know directly how many bikes are at a given station. However a bike id is included in every trip so we can track the bikes and extrapolate. 

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#reading in data
df1 = pd.read_csv('citibike_data/2014-01-CitiBikeTripData.csv') 
df2 = pd.read_csv('citibike_data/2014-02-CitiBikeTripData.csv') 
df3 = pd.read_csv('citibike_data/2014-03-CitiBikeTripData.csv') 
df4 = pd.read_csv('citibike_data/2014-04-CitiBikeTripData.csv') 
df5 = pd.read_csv('citibike_data/2014-05-CitiBikeTripData.csv') 

dataframes=[df1,df2,df3,df4,df5]
for df in dataframes:
    df['starttime']=pd.to_datetime(df['starttime']) #converting time strings into datetime format var
    df['stoptime']=pd.to_datetime(df['stoptime'])   #converting time strings into datetime format var
    
#df=pd.concat([df1,df2,df3,df4,df5])
```

```python
def triptobike(df):
    #create a bikedf dataframe that is indexed by bikeid and starttime.
    bikedf=df[['starttime','stoptime','start station id','end station id','bikeid']].sort('starttime')
    bikedf=bikedf.set_index(['bikeid','starttime'], drop=False)
    bikedf=bikedf.drop(['bikeid'],axis=1)
    bikedf = bikedf.sortlevel(0)
    bikedf.index.names=['bikeid','time']
    return bikedf

def biketowhere(bikedf,startdate,numberofmonths,startrow):
    #converts the bike trip dataframe into the whereisbike datafame
    #which gives a table where rows are date-time hours, columns are bike id number, 
    #and entries are the station number
    
    #hourmarker is a df with hourly time as index,these will be the time markers 
    #where we look to see where the bikes are.
    timerange=pd.date_range(startdate,periods=24*31*numberofmonths,freq='H')
    hourmarker=pd.DataFrame(index=timerange)

    whereisbike=pd.DataFrame(index=timerange)

    for bikeid in bikedf.index.levels[0]:
        #dataframe with the start station info, indexed by start time
        startdf=bikedf.loc[bikeid][['start station id']]
        #dataframe with the end stationo info, indexed by stop time
        stopdf=bikedf.loc[bikeid][['stoptime','end station id']].set_index('stoptime')
        #merge these dataframes
        startstopdf=startdf.join(stopdf,how='outer')
        #change the start station ids to negative numbers so we know it is a start and not a stop.
        #change the start station ids to 0
        startstopdf['start station id']=startstopdf['start station id']-startstopdf['start station id']
        #combine the start and stop columns into one
        ss=pd.concat([startstopdf['start station id'].dropna(),startstopdf['end station id'].dropna()])
        ss.name='station'
        #merge with hourmarker dataframe
        ss=hourmarker.join(ss,how='outer')
        if bikeid in startrow.index:
            ss.station[0]=startrow[bikeid]
        #forwardfill so that time markers now have info from the last stop or start the bike was at 
        ss=ss.fillna(method='ffill')
        #extract out only the hourly times
        whereisbike[str(bikeid)]=hourmarker.join(ss,how='left')

    whereisbike.index.name='datetime'
    whereisbike.rename(columns = lambda x: int(x), inplace=True) #renaming column names to be ints not strings
    return whereisbike

def unknownbikes(whereisbike,datetime):
    #returns the number of unaccounted for bikes at time(row)=datetime
    #these bikes have not yet been logged at a station as of that time.
    return whereisbike.loc[datetime].shape[0]-whereisbike.loc[datetime].count()
    #count doesn't include nans while shape does
    
def stationfill(whereisbike,startdate,numberofmonths):

    #startdate = dt.datetime(2014,5,1,0,0,0)
    timerange=pd.date_range(startdate,periods=24*31*numberofmonths,freq='H')
    hourmarker=pd.DataFrame(index=timerange)

    stationfills=pd.DataFrame()#(index=stationdf.index)
    stationfills.index.name='station id'

    #fills the dataframe with number of bikes at each station at each time
    for datetime in hourmarker.index:
        stationfills[datetime] = whereisbike.loc[datetime].value_counts()
        #if there are no bikes at the station then value_counts give nans, need to fill with 0
    return stationfills.T.fillna(0)
```

```python
bikedf1=triptobike(df1)
nullstartrow=pd.Series(index=bikedf1.index.levels[0])
whereisbike1=biketowhere(bikedf1,dt.datetime(2014,1,1,0),1,nullstartrow)

startrow=whereisbike1.loc[dt.datetime(2014,1,31,23)]
whereisbike2=biketowhere(triptobike(df2),dt.datetime(2014,2,1,0),1,startrow)

startrow=whereisbike2.loc[dt.datetime(2014,2,28,23)]
#startrow=nullstartrow
whereisbike3=biketowhere(triptobike(df3),dt.datetime(2014,3,1,0),1,startrow)

startrow=whereisbike3.loc[dt.datetime(2014,3,31,23)]
#startrow=nullstartrow
whereisbike4=biketowhere(triptobike(df4),dt.datetime(2014,4,1,0),1,startrow)

startrow=whereisbike4.loc[dt.datetime(2014,5,1,0)] #april data spills over onto may1
whereisbike5=biketowhere(triptobike(df5),dt.datetime(2014,5,1,0),1,startrow)
```

```python
#tapping into the citibike json feed
#this will give us the number of total docks per station
#this also gives the current number of available docks/bikes per station (not using that info)
import urllib, json
from pandas.io.json import json_normalize
url = "http://citibikenyc.com/stations/json"

response = urllib.urlopen(url)
data = json.loads(response.read())
stationdf=json_normalize(data['stationBeanList'])
updateTime = data['executionTime']
stationdf['time']=updateTime
stationdf.set_index(['id'],drop=False,inplace=True)
```

```python
startdate = dt.datetime(2014,5,1,0)
stationfills5=stationfill(whereisbike5,startdate,1)
```

```python
#subtracts the number of bikes at each station from the total docks
#gives the number of available docks.
docksfree5=-stationfills5.sub(stationdf['totalDocks'])

#number of stations less than zero docks at some point in the month 
print docksfree5[docksfree5<0].dropna(how='all', axis=1).count().count()
#number of stations less than 2 bikes at some point in the month
print stationfills5[stationfills5<2].dropna(how='all',axis=1).count().count()

#docksfree5.shape
#stationfills5
```
